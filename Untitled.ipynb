```python
#importing gui packages
from tkinter import*
from tkinter import ttk
import tkinter as tk
from tkinter.filedialog import *
import tkinter.messagebox
#importing image handling packages
from PIL import Image,ImageTk
#mporting os handler
import os
#importing hash functions
import hashlib
#importing DES algorithm
from Cryptodome.Cipher import DES
from pyvirtualdisplay import Display

#function for showing error prompt if key is not entered
def pass_alert(i):
    tkinter.messagebox.showinfo("Password Alert","Please enter key {j}.".format(j=i))

def enter():
    if passg.get() == '':
        tkinter.messagebox.showinfo("Value Alert","Please enter key 1")
    elif passg1.get() == '':
        tkinter.messagebox.showinfo("Value Alert","Please enter key 2")
    elif passg2.get() == '':
        tkinter.messagebox.showinfo("Value Alert","Please enter key 3")

#encryption function
def encrypt():
    #creating the file path variable
    global file_path_e
    #getting the passed variables from gui
    enc_pass = passg.get()
    enc_pass1 = passg1.get()
    enc_pass2 = passg2.get()
    #checking if key is empty
    if enc_pass == "":
        pass_alert(1)
    elif enc_pass1 == "":
        pass_alert(2)
    elif enc_pass2 == "":
        pass_alert(3)

    #if key is not empty, encryption process is started
    else:
        #loading the image from os
        filename = tkinter.filedialog.askopenfilename()
        file_path_e = os.path.dirname(filename)

        #generating key of required length using hash algorithm
        #key size for DES is 64 bits including 8 parity bits
        hash=hashlib.shake_128(enc_pass.encode())
        p = hash.digest(8)
        key = p
        hash=hashlib.shake_128(enc_pass1.encode())
        p1 = hash.digest(8)
        key1 = p1
        hash=hashlib.shake_128(enc_pass2.encode())
        p2 = hash.digest(8)
        key2 = p2
        #creating initialization vectors
        iv = p.ljust(8)
        iv1 = p1.ljust(8)
        iv2 = p2.ljust(8)
        print("Encoding key is:\n",key,"\n",key1,"\n",key2)

        #DES Encryption

        #open the image file in read mode and read the data
        input_file = open(filename,'rb')
        input_data = input_file.read()
        input_file.close()

        #Encrypting data using first key
        #creating DES instance in Cipher FeedBack mode
        cfb_cipher = DES.new(key, DES.MODE_CFB, iv)
        #encryption
        enc_data = cfb_cipher.encrypt(input_data)
        #saving encrypted data in file
        enc_file = open(filename, "wb")
        enc_file.write(enc_data)
        enc_file.close()

        #Decrypting data using second key
        input_file = open(filename,'rb')
        input_data = input_file.read()
        input_file.close()
        #creating new DES instance
        cfb_decipher = DES.new(key1, DES.MODE_CFB, iv1)
        #decryption
        plain_data = cfb_decipher.decrypt(input_data)
        #save decrypted data in file
        output_file = open(filename, "wb")
        output_file.write(plain_data)
        output_file.close()

        #Encrypting using third key
        input_file = open(filename,'rb')
        input_data = input_file.read()
        input_file.close()
        #creating new DES instance
        cfb_cipher = DES.new(key2, DES.MODE_CFB, iv2)
        #encryption
        enc_data = cfb_cipher.encrypt(input_data)
        #save encrypted data in file
        enc_file = open(filename, "wb")
        enc_file.write(enc_data)
        enc_file.close()

        #display message box
        tkinter.messagebox.showinfo("Encryption Alert","Encryption ended successfully. File stored as: "+filename+".png")


#decryption function
def decrypt():
    #creating the file path variable
    global file_path_e
    #getting the passed variables from gui
    enc_pass = passg.get()
    enc_pass1 = passg1.get()
    enc_pass2 = passg2.get()
    #checking if key is empty
    if enc_pass == "":
        pass_alert(1)
    elif enc_pass1 == "":
        pass_alert(2)
    elif enc_pass2 == "":
        pass_alert(3)

    #if key is not empty, encryption process is started
    else:
        #loading the image from os
        filename = tkinter.filedialog.askopenfilename()
        file_path_e = os.path.dirname(filename)

        #generating key of required length using hash algorithm
        #key size for DES is 64 bits including 8 parity bits
        hash=hashlib.shake_128(enc_pass.encode())
        p = hash.digest(8)
        key = p
        hash=hashlib.shake_128(enc_pass1.encode())
        p1 = hash.digest(8)
        key1 = p1
        hash=hashlib.shake_128(enc_pass2.encode())
        p2 = hash.digest(8)
        key2 = p2
        iv = p.ljust(8)
        iv1 = p1.ljust(8)
        iv2 = p2.ljust(8)

        #DES Decryption

        #open the image file in read mode and read the data
        input_file = open(filename,'rb')
        input_data = input_file.read()
        input_file.close()

        #Decrypting data using third key
        #creating DES instance in Cipher FeedBack mode
        cfb_decipher = DES.new(key2, DES.MODE_CFB, iv2)
        #decryption
        plain_data = cfb_decipher.decrypt(input_data)
        #save decrypted data in file
        output_file = open(filename, "wb")
        output_file.write(plain_data)
        output_file.close()

        #Encrypting data using second key
        input_file = open(filename,'rb')
        input_data = input_file.read()
        input_file.close()
        #creating new DES instance
        cfb_cipher = DES.new(key1, DES.MODE_CFB, iv1)
        #encryption
        enc_data = cfb_cipher.encrypt(input_data)
        #saving encrypted data in file
        enc_file = open(filename, "wb")
        enc_file.write(enc_data)
        enc_file.close()

        #Decrypting data using first key
        input_file = open(filename,'rb')
        input_data = input_file.read()
        input_file.close()
        #creating DES instance
        cfb_decipher = DES.new(key, DES.MODE_CFB, iv)
        #decryption
        plain_data = cfb_decipher.decrypt(input_data)
        #saving decrypted data in file
        output_file = open(filename, "wb")
        output_file.write(plain_data)
        output_file.close()

        #display message box
        tkinter.messagebox.showinfo("Decryption Alert","Decryption ended successfully File Stored as: "+filename+".png")


# GUI

#creating tkinter instance
display = Display(visible=0, size=(800, 600))
display.start()
top=tk.Tk()
#setting window dimensions
top.geometry("500x600")
top.resizable(0,0)
#set title bar
top.title("ImageEncryption")

#set title in window
title="Image Encryption Using DES"
msgtitle=Message(top,text=title)
msgtitle.config(font=('helvetica',17,'bold'),width=200)
msgtitle.pack()

#insert seperator
sp="---------------------------------------------------------------------"
sp_title=Message(top,text=sp)
sp_title.config(font=('arial',12),width=650)
sp_title.pack()

#label for text box for key 1
passlabel = Label(top, text="Enter Encryption/Decryption Key 1:")
passlabel.pack()
#text box for key 1
passg = Entry(top, show="*", width=20)
passg.config(highlightthickness=1,highlightbackground="blue")
passg.pack()

#lael for text box for key 2
passlabel1 = Label(top, text="Enter Encryption/Decryption Key 2:")
passlabel1.pack()
#text box for key 2
passg1 = Entry(top, show="*", width=20)
passg1.config(highlightthickness=1,highlightbackground="blue")
passg1.pack()

#label for text box for key 3
passlabel2 = Label(top, text="Enter Encryption/Decryption Key 3:")
passlabel2.pack()
#text box for key 3
passg2 = Entry(top, show="*", width=20)
passg2.config(highlightthickness=1,highlightbackground="blue")
passg2.pack()

#insert 'Encrypt' button with action 'encrypt'
encrypt=Button(top,text="Encrypt",width=28,height=3,command=encrypt)
encrypt.pack(side=LEFT)

#insert 'Decrypt' button with action 'decrypt'
decrypt=Button(top,text="Decrypt",width=28,height=3,command=decrypt)
decrypt.pack(side=RIGHT)

#execute application
top.mainloop()
```
